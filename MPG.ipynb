```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

```

###CODE


```python

# Example dataset: Car weight (1000s of pounds) vs MPG (miles per gallon)
# Data from a simple illustrative example
weights = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.0]).reshape(-1, 1)  # Feature (X)
mpg = np.array([30, 28, 25, 22, 20, 17])  # Label (Y)

# Create and train linear regression model
model = LinearRegression()
model.fit(weights, mpg)

# Predictions
mpg_pred = model.predict(weights)

# Model parameters
bias = model.intercept_
weight = model.coef_[0]

# Metrics
mse = mean_squared_error(mpg, mpg_pred)
r2 = r2_score(mpg, mpg_pred)

# Plot
plt.scatter(weights, mpg, color="blue", label="Actual data")
plt.plot(weights, mpg_pred, color="red", label="Fitted line")
plt.xlabel("Car Weight (1000 lbs)")
plt.ylabel("Miles per Gallon (MPG)")
plt.title("Linear Regression: MPG vs Weight")
plt.legend()
plt.show()

(bias, weight, mse, r2)
```
